## Part 1: Working with JSON Data from GitHub
Raw JSON -> python list ->  Dataframe -> CSV


```python
# necessary imports
import requests
import pandas as pd
```

```python
# get raw link
url = "https://raw.githubusercontent.com/Kimaiyoo/data_engineering/refs/heads/main/python/movie_json/movie_inventory.json"

# send request
response = requests.get(url)

# display response status code
response.status_code
```

```python
# convert  json to python list
movies= response.json()
movies
```

```python
# check data type
type(movies)
```

```python
# convert data to dataframe
movies_df = pd.DataFrame(movies)

# display sample data
movies_df.head()
```

```python
# save dataframe as csv
movies_df.to_csv("movies.csv", index= False)
print("Successfully saved movies.csv")
```

## Part 2: Working with DummyJSON API Endpoints
Products endpoint
https://dummyjson.com/products

Carts endpoint
https://dummyjson.com/carts

```python
# products endpoint
products_url= "https://dummyjson.com/products"

res = requests.get(products_url)
res.status_code
```

```python
products = res.json()
type(products)
```

```python
products.keys()
```

```python
products['products'][:4]
```

```python
type(products['products'])
```

the products list contains nested dicts so next step : flatten and normalize

```python
products = products['products']
```

```python
# normalize at top level
products_df = pd.json_normalize(products)
products_df.head()
```

```python
# explode reviews column
products_exploded = products_df.explode('reviews').reset_index(drop=True)
products_exploded.head()
```

```python
products_df.info() # still remains the same
```

```python
# flatten dictionaries inside reviews column
if not products_exploded['reviews'].isna().all():
    reviews_products = pd.json_normalize(products_exploded['reviews'])

    # Add a prefix so you know these columns belong to the review
    reviews_products = reviews_products.add_prefix('review_')

    # Join the exploded product data with the flattened review data(drop old review column)
    products_final = pd.concat([products_exploded.drop(columns=['reviews']), reviews_products], axis=1)
else:
    # If no reviews existed at all
    products_final = products_exploded.drop(columns=['reviews'])
```

```python
products_final.info() # reviews column dropped and its data flattened to form new cols
```

```python
# save as csv
products_final.to_csv('products.csv', index=False)
```

```python
# carts
carts_url= "https://dummyjson.com/carts"

res = requests.get(carts_url)
res.status_code
```

```python
carts = res.json()
type(carts)
```

```python
carts.keys()
```

```python
carts['carts'][:4]
```

```python
carts = carts['carts']
```

```python
# normalize at top level
carts_df = pd.json_normalize(carts)
carts_df.head()
```

```python
# rename to avoid confusion later
carts_df = carts_df.rename(columns={
    'id': 'cart_id',
    'total': 'cart_total',
    'discountedTotal': 'cart_discountedTotal'
})
```

```python
carts_df.head()
```

```python
# Explode the list of products
carts_exploded = carts_df.explode('products').reset_index(drop=True)
carts_exploded.info()
```

```python
# Normalize the dictionaries inside the 'products' column
products_df = pd.json_normalize(carts_exploded['products'])

# Add a prefix to specify it's product details
products_df = products_df.add_prefix('prod_')

# join them
carts_final = pd.concat([carts_exploded.drop(columns=['products']), products_df], axis=1)
carts_final.info()
```

```python
# create csv
carts_final.to_csv("carts.csv", index=False)
print("Successfully saved as carts.csv")
```
