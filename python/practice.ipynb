Question: A principal checking report cards: For EACH student, go through EACH subject and write the score. Then calculate the average for that student. Move to the next student.
 
subjects = ["Maths", "English", "Science"]
scores = [
    [78, 85, 90],   # Amina
    [65, 72, 68],   # Brian
    [91, 88, 95]    # Njeri
]

```python
students = ["Amina", "Brian", "Njeri"]
subjects = ["Maths", "English", "Science"]
scores = [
    [78, 85, 90],   # Amina
    [65, 72, 68],   # Brian
    [91, 88, 95]    # Njeri
]


for s_i, student in enumerate(students):
    print(f"\nReport card for {student}")
    # start sum with 0
    total = 0
    for sub_i, subject in enumerate(subjects):
        # get the scores for current student
        score = scores[s_i][sub_i]
        # add score to prev total
        total += score
        print(f"{subject} : {score}")
    
    # calculate average
    average = total/ len(subjects)
    print(f"Average score: {average:.2f}")
        

```

```python
# Q7.  Write a function summarise(numbers) that takes a list and returns: count, total, minimum, 
#maximum, and average — all as separate return values. Test with [45, 78, 62, 91, 55, 70]. 
def summarise(numbers):
    count = len(numbers)
    total = sum(numbers)
    minimum = min(numbers)
    maximum= max(numbers)
    average = total / count

    return f"Count: {count}\nTotal: {total}\nMinimum: {minimum}\nMaximum: {maximum}\nAverage: {average:.2f}"

test = [45, 78, 62, 91, 55, 70]
print(summarise(test))
 
```

```python
# Q1.  Write a function safe_divide(a, b) that divides a by b. Catch ZeroDivisionError and TypeError separately.
#  Return None for both error cases. Test with: (10, 2), (10, 0), (10, 'five').
def safe_divide(a, b):
    try:
        quo = a / b
        return quo
    except ZeroDivisionError:
        return None
    except TypeError:
        return None
       
   

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide(10, 'five'))
```

```python
# Q2.  Write a function ask_age() that keeps asking the user for their age until they enter a valid whole number between 1 and 120. 
# Use a while True loop with try/except.
def ask_age():
    while True  :
        try:
            age = int(input("How old are you?: "))
            if age < 0:
                continue
            elif age < 0 or age >120:
                continue
            else:
                return f"You are {age} years old."

        except ValueError:
                print("Age should be a whole number.")
            
        
ask_age()
```

```python
# Q3.  Given a list: ['4', '7', '', '2', 'N/A', '9', '0']. Loop through and try to convert each to int. 
# Skip empty strings and non-numeric values. Collect valid numbers in one list and error messages in another.
#  Print both lists and the total of valid numbers.
nums = ['4', '7', '', '2', 'N/A', '9', '0']

def to_int(numbers):
    valid = []
    errors = []
    for i,n in enumerate(numbers):
        
        try:
            n = int(n)
            valid.append(n)
            
        except Exception as e:
            errors.append(f"Error at index {i} - {e}")

    total_valid = sum(valid)
    results = f"Valid list: {valid}\n\nTotal of valid: {total_valid}\n\nErrors: {errors}"

    return  results

print(to_int(nums))

```

```python
nums = [5, 8, 9]
def get_highest_score(nums):
    # a check for empty list
    if not nums:
        return None
    
    highest = nums[0] # initialize first element as highest

    # iterate through list to find larger value
    for n in nums:
        if n > highest:
            highest= n

    return highest

print(get_highest_score([55, 90, 73, 88]))
print(get_highest_score([]))
```
